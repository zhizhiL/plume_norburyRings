```python
import numpy as np
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
```

```python
frame = np.load('lineplume_sims_3D/temp/frame_66.npy')
x, y, z = frame[:, 1:4].T

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='b', marker='o')

# Set axis
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-6, 2)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# set viewing angle
ax.view_init(elev=5, azim=270)

# Show plot
plt.show()
```

```python
# Function to load frame data
def load_frame(frame_number):
    return np.load(f'lineplume_sims_3D/temp/frame_{frame_number}.npy')

# Function to update the plot for each frame
def update_plot(frame_number, scatter):
    frame = load_frame(frame_number)
    x, y, z = frame[:, 1:4].T
    scatter._offsets3d = (x, y, z)
    return scatter,

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Set axis limits
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-6, 2)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set viewing angle
ax.view_init(elev=5, azim=270)

# Initialize the scatter plot
frame = load_frame(1)
x, y, z = frame[:, 1:4].T
scatter = ax.scatter(x, y, z, c='b', marker='o')

# Create the animation
ani = animation.FuncAnimation(fig, update_plot, frames=range(1, 268), fargs=(scatter,), interval=200)

# Save the animation as a movie file
ani.save('3d_scatter_movie.mp4', writer='ffmpeg', dpi=300)

# Show the plot
plt.show()
```
